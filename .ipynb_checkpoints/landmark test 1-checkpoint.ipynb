```python
import cv2
import mediapipe as mp

from pyfirmata import Arduino ,SERVO,util
from time import sleep

port='COM4'
board=Arduino(port)
```

```python
#finger 3
#middle finger
#130
pin=9
board.digital[pin].mode=SERVO

#finger 4
#wedha finger
#130
pin2=10
board.digital[pin2].mode=SERVO

#finger 5
#sula finger
#120
pin3=11
board.digital[pin3].mode=SERVO

#finger 2
#dabara finger
pin4=6
board.digital[pin4].mode=SERVO
```

```python
usingpin=5

def rotate(pin,angle):
    board.digital[pin].write(angle)
    sleep(0.002)

def get_finger_status(finger_landmarks, finger_name):
    # Calculate the distance between the two landmarks
    dist = abs(finger_landmarks[0].y - finger_landmarks[2].y)

    # Determine the finger status based on the distance
    if dist < 0.03:
        
        if finger_name=="Thumb":
            return f"{finger_name}: removed"
        
        if finger_name=="Finger 2":
            usingpin=6
            angle=60
        if finger_name=="Finger 3":
            usingpin=9
            angle=160
        if finger_name=="Finger 4":
            usingpin=10
            angle=100
        if finger_name=="Finger 5":
            usingpin=11
            angle=90
        rotate(usingpin,angle)            
        return f"{finger_name}: Middle"
    elif finger_landmarks[0].y > finger_landmarks[2].y:
        
        angle=0
        
        if finger_name=="Thumb":
            return f"{finger_name}: removed"
        
        if finger_name=="Finger 2":
            usingpin=6
            angle=180
        if finger_name=="Finger 3":
            usingpin=9
        if finger_name=="Finger 4":
            usingpin=10
            
        if finger_name=="Finger 5":
            usingpin=11
            angle=180
        rotate(usingpin,angle)
        
        return f"{finger_name}: Up"
    else:
        
        if finger_name=="Thumb":
            return f"{finger_name}: removed"
        
        if finger_name=="Finger 2":
            usingpin=6
            angle=60
        if finger_name=="Finger 3":
            usingpin=9
            angle=140
        if finger_name=="Finger 4":
            usingpin=10
            angle=130
        if finger_name=="Finger 5":
            usingpin=11
            angle=60
        rotate(usingpin,angle)
        
        return f"{finger_name}: Closed"

def main():
    mp_drawing = mp.solutions.drawing_utils
    mp_hands = mp.solutions.hands

    # Initialize webcam
    cap = cv2.VideoCapture(0)

    # Initialize hand tracking
    with mp_hands.Hands(
        static_image_mode=False,
        max_num_hands=1,
        min_detection_confidence=0.5) as hands:

        while cap.isOpened():
            success, image = cap.read()
            if not success:
                print("Ignoring empty camera frame.")
                continue

            # Convert the image to RGB and process it with mediapipe
            image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
            results = hands.process(image)

            # Draw hand landmarks on the image
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    mp_drawing.draw_landmarks(
                        image, hand_landmarks, mp_hands.HAND_CONNECTIONS)

                    # Print finger status and landmarks
                    if hand_landmarks.landmark:
                        for i, landmark in enumerate(hand_landmarks.landmark):
                            # Calculate landmark coordinates on the screen
                            height, width, _ = image.shape
                            cx, cy = int(landmark.x * width), int(landmark.y * height)

                            # Display finger landmarks and status on the screen
                            if i in [4, 8, 12, 16, 20]:
                                finger_name = "Thumb" if i == 4 else f"Finger {((i - 4) // 4) + 1}"
                                finger_landmarks = hand_landmarks.landmark[i-3:i+1]
                                finger_status = get_finger_status(finger_landmarks, finger_name)
                                cv2.putText(image, finger_status,
                                            (cx, cy), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

            cv2.imshow('Hand Tracking', image)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == '__main__':
    main()
```
